```python
import math
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score 

from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.metrics import roc_curve, auc, roc_auc_score

from sklearn.neighbors import KNeighborsClassifier

from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import roc_curve, auc, roc_auc_score
```

```python
#Importing the dataset 
df=pd.read_csv("/home/umair/Desktop/Data Science and BI/data mining/occupancy_data/DataCleaned.csv")
test=pd.read_csv("/home/umair/Desktop/Data Science and BI/data mining/occupancy_data/DataCleanedTest.csv")
test2=pd.read_csv("/home/umair/Desktop/Data Science and BI/data mining/occupancy_data/DataCleanedTest2.csv")
```

```python
#Drop the useless column
df.drop(['Unnamed: 0'], axis=1, inplace =True)
df.drop(['date'], axis=1, inplace =True)
df.drop(['DayName'], axis=1, inplace =True)

test.drop(['Unnamed: 0'], axis=1, inplace =True)
test.drop(['date'], axis=1, inplace =True)
test.drop(['DayName'], axis=1, inplace =True)

test2.drop(['Unnamed: 0'], axis=1, inplace =True)
test2.drop(['date'], axis=1, inplace =True)
test2.drop(['DayName'], axis=1, inplace =True)
```

```python
#preparing the two datasets splitting the attributes with the class 
attributes = [col for col in df.columns if 
              col == 'IsWorkDay' 
              or col == 'Hour' 
              or col == 'Light'
              or col == 'IsWorkHour'
              or col == 'Temperature' 
              or col == 'Humidity'
              or col == 'CO2'
             ]
X_train = df[attributes]
y_train = df['Occupancy']

attributes = [col for col in test.columns if 
              col == 'IsWorkDay' 
              or col == 'Hour' 
              or col == 'Light'
              or col == 'IsWorkHour'
              or col == 'Temperature' 
              or col == 'Humidity'
              or col == 'CO2'
             ]
X_test = test[attributes]
y_test = test['Occupancy']

attributes = [col for col in test2.columns if 
              col == 'IsWorkDay' 
              or col == 'Hour' 
              or col == 'Light'
              or col == 'IsWorkHour'
              or col == 'Temperature' 
              or col == 'Humidity'
              or col == 'CO2'
             ]
X_test2 = test2[attributes]
y_test2 = test2['Occupancy']
```

```python
clf = KNeighborsClassifier(n_neighbors=81, weights='uniform')
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_predB = clf.predict_proba(X_test)

print('Accuracy %s' % accuracy_score(y_test, y_pred))
print('F1-score %s' % f1_score(y_test, y_pred, average=None))
print(classification_report(y_test, y_pred))
```

```python
y_pred2 = clf.predict(X_test2)
y_predB2 = clf.predict_proba(X_test2)

print('Accuracy %s' % accuracy_score(y_test2, y_pred2))
print('F1-score %s' % f1_score(y_test2, y_pred2, average=None))
print(classification_report(y_test2, y_pred2))
```

```python
#trying the K-NN algorithm with K=78 which is the best k!!
from sklearn.metrics import confusion_matrix
clf = KNeighborsClassifier(n_neighbors=78, weights='uniform')
clf.fit(X_train, y_train)

y_predB = clf.predict(X_test)

print('Accuracy %s' % accuracy_score(y_test, y_predB))
print('F1-score %s' % f1_score(y_test, y_predB, average=None))
print(classification_report(y_test, y_predB))
print(confusion_matrix(y_test, y_predB))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import scikitplot as skplt

features_combs_list = [
    ('IsWorkDay', 'Hour', 'IsWorkHour', 'Temperature', 'Humidity', 'Light', 'CO2')
]

hyper_params_space = [
    {
        'n_neighbors': np.arange(78, 82),
    },
]

for features in features_combs_list:
    print(features)
    print('===================================')
    X = X_train.loc[:, features]
    X_t1 = X_test.loc[:, features]
    X_t2 = X_test2.loc[:, features]

    knn = GridSearchCV(KNeighborsClassifier(), hyper_params_space,
                       scoring='accuracy')
    knn.fit(X, y_train)
    
    print('Best parameters set:')
    print(knn.best_params_)
    print()
    
    preds = [
        (knn.predict(X), y_train, 'Train', knn.predict_proba(X)),
        (knn.predict(X_t1), y_test, 'Test1', knn.predict_proba(X_t1)),
        (knn.predict(X_t2), y_test2, 'Test2', knn.predict_proba(X_t2))
    ]
    
    for pred in preds:
        print(pred[2] + ' Classification Report:')
        print()
        print('Accuracy %s' % accuracy_score(pred[1], pred[0]))
        print()
        print(classification_report(pred[1], pred[0]))
        print()
        print(pred[2] + ' Confusion Matrix:')
        print(confusion_matrix(pred[1], pred[0]))
        print()
        y_predB = clf.predict_proba(X_t1)
        lb = LabelBinarizer()
        lb.fit(y_test)
        lb.classes_.tolist()
        fpr, tpr, _ = roc_curve(pred[1], pred[0])
        roc_auc = auc(fpr, tpr)
        print(roc_auc)
        roc_auc = roc_auc_score(pred[1], pred[0], average=None)
        plt.figure(figsize=(8, 5))
        plt.title(pred[2] +' ROC Curve')
        plt.plot(fpr, tpr, label='ROC curve (area = %0.5f)' % roc_auc)

        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlim([0.0, 1])
        plt.ylim([0.0, 1])
        plt.xlabel('False Positive Rate', fontsize=20)
        plt.ylabel('True Positive Rate', fontsize=20) 
        plt.tick_params(axis='both', which='major', labelsize=22)
        plt.legend(loc="lower right", fontsize=14, frameon=False)
        plt.show()
        
        skplt.metrics.plot_cumulative_gain(pred[1], pred[3])
        plt.title(pred[2] +' Lift Chart')
        plt.show()
```
