```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import LocalOutlierFactor

from collections import defaultdict
```

```python
df=pd.read_csv("../datatraining.csv", sep=";")
test=pd.read_csv("../datatest.csv", sep=";")
```

```python
class_name='Occupancy'
attributes = [col for col in df.columns if 
              col == 'Temperature'
              or col == 'Humidity'
              or col == 'Light'
              or col == 'CO2'
             ]
X_train = df[attributes]
y_train = df[class_name]

X_test=test[attributes]
y_test = df[class_name]
```

```python
X_trainv=X_train.values
X_testv=X_test.values
```

```python
clf = LocalOutlierFactor(n_neighbors=10)
outliers = clf.fit_predict(X_train)
```

```python
np.unique(outliers, return_counts=True)
```

```python
np.where(outliers==-1)
```

```python
plt.scatter(X_train['Light'], X_train['CO2'])
plt.scatter(X_train['Light'].loc[np.where(outliers==-1)], X_train['CO2'].loc[np.where(outliers==-1)], color='red')
plt.xlabel('Light')
plt.ylabel('CO2') 
plt.show()
```

```python
plt.scatter(X_train['Temperature'], X_train['Humidity'])
plt.scatter(X_train['Temperature'].loc[np.where(outliers==-1)], X_train['Humidity'].loc[np.where(outliers==-1)], color='red')
plt.xlabel('Temperature')
plt.ylabel('Humidity') 
plt.show()
```

```python
plt.scatter(X_train['Light'], X_train['Humidity'])
plt.scatter(X_train['Light'].loc[np.where(outliers==-1)], X_train['Humidity'].loc[np.where(outliers==-1)], color='red')
plt.xlabel('Light')
plt.ylabel('Humidity') 
plt.show()
```

```python
plt.scatter(X_train['Light'], X_train['Temperature'])
plt.scatter(X_train['Light'].loc[np.where(outliers==-1)], X_train['Temperature'].loc[np.where(outliers==-1)], color='red')
plt.xlabel('Light')
plt.ylabel('Temperature') 
plt.show()
```

```python
np.max(clf.negative_outlier_factor_[np.where(outliers==-1)])
```
